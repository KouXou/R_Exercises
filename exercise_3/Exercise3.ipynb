```r
# Function to plot Data
plot_data <- function(in_data) {
  pairs(in_data)
  summary(in_data)
}
```

## Ερώτημα i

Θεωρητικό υπόβαθρο Πολλαπλης Γραμμικής Παλινδρόμησης

Όταν μία μεταβητή Υ εξαρτάται από 2 ή περισσότερες ανεξάρτητες μεταβλητές X₁,X₂,...Xₚ χρησιμοποιούμε το γραμμικό μοντέλο

Yₑ = α + β₁X₁ + β₂X₂ + … + βₚXₚ+ε

Για να διερευνηθεί η σχέση μεταξύ της Υ και των X₁,X₂,...Xₚ παίρνουμε δείγμα μεγέθους n και για κάθε άτομο του δείγματος καταγράφονται οι τιμές των συγκεκριμένων μεταβλητών.

Τα σφάλματα θεωρούνται ανεξάρτητες τ.μ. από την N(0,1) ενώ οι ερμηνευτικές μεταβλητές X₁,X₂,...Xₚ , όπως και στο απλό γραμμικό μοντέλο, δεν θεωρούνται τυχαίες.

Στο πολλαπλό μοντέλο θεωρούμε ότι τα σημεία δεν βρίσκονται «κοντά» σε μια ευθεία αλλά «κοντά» σε ένα (υπερ)επίπεδο στις n διαστάσεις.

## Ερώτημα iv

Στη συνέχεια υπολογίζουμε τα γραμμικά μοντέλα παλινδρόμησης του maxO3 σε σχέση με Τ9, Τ12 και Τ15 αντίστοιχα.

```r
maxO3 <- ozone_data[,1]
T9 <- ozone_data[,2]
T12 <- ozone_data[,3]
T15 <- ozone_data[,4]
lm1=lm(maxO3~T9,data=ozone_data) 
lm2=lm(maxO3~T12,data=ozone_data) 
lm3=lm(maxO3~T15,data=ozone_data) 
lm1
lm2
lm3
```

Οι παράμετροι για το Τ12 είναι -27.420 και 5.469 .

Για το μοντέλο maxO3~Τ12 υπολογίζω το συντελεστή προσδιορισμού και τον προσαρμοσμένο συντελεστή προσδιορισμού



```r
summary(lm2)
print(paste('Συντελεστής Προσδιορισμού : ', summary(lm2)$r.squared ))
print(paste('Προσαρμοσμένος Συντελεστής Προσδιορισμού : 0.6116 '))
```

Βλέπουμε ότι ο συντελεστής προσδιορισμού είναι 0.615 που σημαίνει ότι η γραμμική σχέση εξάρτησης των μεταβλητών 
μας είναι ισχυρή, υπό την προϋπόθεση ότι το γραμμικό μοντέλο είναι το κατάλληλο.

## Ερώτημα v

Με χρήση της συνάρτησης abline σχεδιάζουμε την ευθεία παλινδρόμησης. 

```r
plot(maxO3 ~ T12)
abline(lm2)
```

Με τη χρήση της rstudent() βρίσκουμε τα studentised υπόλοιπα, τα οποία έχουν την ίδια διακύμανση.

```r
res<--rstudent(lm2)
plot(res,pch=15,cex=.5,ylab="Residuals")
abline(h=c(-2,0,2),lty=c(2,1,2))
```

Μόνο 4 studentised υπόλοιπα βρίσκονται εκτός του ορίου [-2,2] πράγμα που ακολουθεί τη θεωρία ότι το 95% των υπολοίπων βρίσκεται σε αυτό το διάστημα.

Στη συνέχεια χρησιμοποιούμε την εντολή regsubsets() απο το πακέτο leaps.

```r
leaps<-regsubsets(maxO3 ~ T9 + T12 + T15 + Ne9 + Ne12 + Ne15 + Wx9 + Wx12 + Wx15 + maxO3y
                  , data=ozone_data, nbest=1)
plot(leaps, scale="adjr2")
```

Από το διάγραμμα που προκύπτει φαίνεται ότι ένα μοντέλο με τα intercept, Τ12, Ne9, Wx9 και maxO3y θα έχει adjusted R-squared περίπου 0.75 .
