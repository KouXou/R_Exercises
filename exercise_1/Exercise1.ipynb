# Άσκηση 1

Έστω **Μ** το ενδεχόμενο να είναι κάποιος μεταπτυχιακός φοιτητής

Έστω **Π** το ενδεχόμενο να είναι κάποιος προπτυχιακός φοιτητής

Έστω **Κ** το ενδεχόμενο να είναι κάποιος φοιτητής καπνιστής

**P(M)=0.2**

**P(Π)=0,8**

**P(Κ/Μ)=0,23**

**P(K/Π)=0,15**

Ψάχνουμε το P(M/K) 

Από το θεώρημα της ολικής πιθανότητας έχουμε :

**P(K)=P(K/M)*P(M)+P(K/Π)*P(Π)=0,23*0,2+0,15*0,8=0,166**

Σύμφωνα με τον τύπο του Bayes έχουμε:

**P(M/K)=P(K/M)*P(M)/P(K)=0.23*0.2/0.166=0.277**

**Άρα το P(M/K)=0.277**

## Ερώτημα i

Σύμφωνα με τα όσα έχουν αναλυθεί ένας τυχαία επιλεγμένος φοιτητής του κολλεγίου είναι πιο πιθανό να είναι

**προπτυχιακός αφού P(Π)=0,8**

## Ερώτημα ii

Με βάση το ερώτημα έχουμε 

**P(Π/Κ)=1-P(M/K)=1-0.277=0.723**

**Άρα η πιθανότητα P(Π/Κ)=0,723**

## Ερώτημα iii

Με βάση το ερώτημα έχουμε:

**P(E/M)=0.3**

**P(E/Π)=0,1**

Θεωρούμε τυχαία μεταβλητή **Χ1** για το πεδίο Καπνιστής με τιμές **ΝΑΙ/ΟΧΙ**

Θεωρούμε τυχαία μεταβλητή **Χ2** για το πεδίο Ζεί στην εστία με τιμές **ΝΑΙ/ΟΧΙ**

Θεωρούμε τυχαία μεταβλητή **Υ** για το πεδίο Φοιτητής με τις εξής τιμές **Προπτυχιακός/Μεταπτυχιακός**

Σύμφωνα με τον τύπο του Bayes έχουμε:

**P(Y=Μεταπτυχιακός|Χ1=χ1, Χ2=χ2)=P(X1=x1,X2=x2|Y=Μεταπτυχιακός)*P(Y=Μεταπτυχιακός)/P(X1=x1, X2=x2)
(1)**

Και

**P(Y=Προπτυχιακός|Χ1=χ1, Χ2=χ2)=P(X1=x1,X2=x2|Y=Προπτυχιακός)*P(Y=Προπτυχιακός)/P(X1=x1, X2=x2)
(2)**

Θεωρούμε ότι οι τυχαίες μεταβλητές είναι υπό συνθήκη ανεξάρτητες δοθείσα τιμή  του Υ, Δηλαδή εάν γνωρίζουμε την τιμή της τυχαίας μεταβλητής Χ1 με δεδομένο ότι γνωρίζουμε την τιμή της Υ δεν μπορούμε να εξάγουμε κανένα συμπέρασμα αναφορικά με την τιμή της τυχαίας μεταβλητής Χ2 και συνεπώς ισχύει η σχέση:

**P(X1=x1, X2=x2|Y=Μεταπτυχιακός= P(X1=x1|Y=Μεταπτυχιακός)*P(X2=x2|Μεταπτυχιακός)=0,23*0,3=0,069**

**P(X1=x1, X2=x2|Y=Προπτυχιακός= P(X1=x1|Y=Προπτυχιακός)*P(X2=x2|Προπτυχιακός)=0,15*0,1=0,015**

Έτσι με βάση τα ανωτέρω η πιθανότητα (1) διαμορφώνετε ως εξής:

**P(X1=x1,X2=x2|Y=Μεταπτυχιακός)*P(Y=Μεταπτυχιακός)=0,069*0,2=0,0138**

Και η πιθανότητα (2)

**P(X1=x1,X2=x2|Y=Προπτυχιακός)*P(Y=Προπτυχιακός)=0,015*0,8=0,012**

Ο παρονομαστής  **P(X1=x1, X2=x2)** επειδή είναι κοινός και στην 1 και στην 2 δεν λαμβάνετε υπόψη.

Έτσι έχουμε 

**P(Y=Μεταπτυχιακός|Χ1=χ1, Χ2=χ2)= 0,0138**

Και

**P(Y=Προπτυχιακός|Χ1=χ1, Χ2=χ2)= 0,012**
